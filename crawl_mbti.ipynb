```python
data = pd.read_csv('/kaggle/input/quotes/quotes.csv')
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def search(query):
    base_url = "https://personalityatwork.co/celebrity/profile/"
    formatted_query = query.lower().replace(" ", "-")
    url = base_url + formatted_query
    response = requests.get(url)
    if response.status_code == 200:
        # Sử dụng BeautifulSoup để phân tích nội dung HTML
        soup = BeautifulSoup(response.text, 'html.parser')
    
        # Tìm tất cả các thẻ h3 trong trang web
        h3_tags = soup.find_all('h3')
        result = []
        # In nội dung của các thẻ h3
        for tag in h3_tags:
            result.append(tag.text)  # hoặc có thể lưu vào danh sách hoặc làm việc khác với nội dung này
    else:
        print('Yêu cầu không thành công')
        result = 0
    return result
```

```python
mbti = []
for i in range(len(data)):
    r = search(data['Column1'][i])
    mbti.append(r)
    print(i)
    print(r[0])
```

```python
data['mbti'] = mbti
```

```python
for i in range(len(data)):
    data['mbti'][i] = mbti[i][0]
    print(i)
```

```python
df = data[data['mbti'] != 'Self-awareness']
```
